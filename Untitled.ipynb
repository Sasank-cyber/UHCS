```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("./data/sentimentdataset.csv")
c = df.columns.tolist()
df
```

```python
df["text_lenght"] = df["Text"].apply(len)
df["text_exclamation"] = df["Text"].str.count("!")
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("text", TfidfVectorizer(stop_words="english"), "Text"),
        ("num", MinMaxScaler(), ["text_lenght", "text_exclamation"])
    ]
    
)
```

```python
model = Pipeline([
    ("features", preprocessor),
    ("nb", MultinomialNB(alpha=1.0))
])
```

```python
columns = ["Text","text_lenght","text_exclamation"]
x = df[columns]
y = df["Sentiment"]
x
```

```python
train_x, val_x, train_y, val_y = train_test_split(x,y,test_size=0.2)
```

```python
model.fit(train_x,train_y)
```

```python
y_pred = model.predict(val_x)
```

```python
user_text = "The service was absolutely terrible!!!"

test_input = pd.DataFrame([{
    "Text": user_text,
    "text_lenght": len(user_text),
    "text_exclamation": user_text.count("!")
}])
```

```python
prediction = model.predict(test_input)
print("Predicted sentiment:", prediction[0])
```
